## Automation example

```python
import random
import time
from tqdm.auto import tqdm

def scrape_once():
    response = requests.get('https://www.midjourney.com/api/app/recent-jobs/', params=params, cookies=cookies, headers=headers)
    response.raise_for_status()
    return response.json()

responses = []
for i in tqdm(range(10)):
    time.sleep(1 + random.randint(3, 6))
    data = scrape_once()
    print(len(data))
    responses.append(data)
```

```python
import itertools

data = pd.json_normalize(
    itertools.chain.from_iterable(responses))
```

```python
data.info()
```

```python
data.sample(30)
```

## Pros and cons

### Pros
- less traffic
- usually, less security barriers
- parsing is trivial and fast
- less changes over time
### Cons
- Not every website has a public API
- No documentation
- Some websites are sending rendered HMTL pages instead of JSON
- Often requires a cookie. You might need to program automatic cookie setting
